```python
def area(radius):
    return (3.14 * radius * radius)

radius = float(input("Enter the radius:"))
               
result = area(radius)
print("Area of the circle is:", result)
```

```python
def largestnum(a, b, c):
    return max(a, b, c)
a = int(input("First number: "))
b = int(input("Second number: "))
c = int(input("Third number: "))

print("The largest number is:", largestnum(a, b, c))
```

```python
def stringlength(text):
    count = 0
    for character in text:
        count += 1
    return count

input = input("Enter a string: ")
print("Length of string is:", stringlength(input))
```

```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("It is not a prime number")
            break
    else:
        print("It is a prime number")
else:
    print("It is not a prime number")
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
n = int(input("Enter a number: "))
print("Fibonacci number is:", fibonacci(n))
```

```python
def reversestring(s):
    if len(s) <= 1:
        return s
    else:
        return reversestring(s[1:]) +s[0]
```

```python
n = int(input("Enter a number: "))

total = 0

for i in range(1, n + 1):
    total = total + i

print("Sum from 1 to", n, "is:", total)
```

```python
numbers = [1, 2, 3, 3, 5, 6, 8, 10]

count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("Number of even numbers:", count)
```
