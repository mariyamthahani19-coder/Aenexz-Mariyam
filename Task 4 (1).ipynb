```python
import pandas as pd

df = pd.read_csv("mtsamples.csv.zip")   
df.head()
```

```python
df.info()
df.shape
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['transcription'])
```

```python
df['medical_specialty'] = df['medical_specialty'].fillna("Unknown")
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
import re

def clean_text(text):
    text = text.lower()                      
    text = re.sub(r'\d+', '', text)           
    text = re.sub(r'[^\w\s]', '', text)       
    text = re.sub(r'\s+', ' ', text).strip()  
    return text

df['clean_transcription'] = df['transcription'].apply(clean_text)
```

```python
df['medical_specialty'].value_counts()
```

```python
df['medical_specialty'] = df['medical_specialty'].replace({
    "Cardiovascular / Pulmonary": "Cardiology",
    "Orthopedic": "Orthopedics"
})
```

```python
df.to_csv("cleaned_mtsamples.csv", index=False)
```
